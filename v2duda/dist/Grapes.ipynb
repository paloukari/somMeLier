```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/database.csv')
df
```

```python
grapes_parsed=[]
for grape in df["grapes"]:
    if grape[0]==' ':
        grapes_parsed.append(grape[1:].replace('\n',''))
    else:
        grapes_parsed.append(grape.replace('\n',''))
```

```python
df["grapes"]=grapes_parsed
```

```python
db=df[["types","grapes"]].copy()
db
```

```python
db["count"]=1
db
```

```python
db["grapes"]=[x.split(',') for x in db["grapes"]]
db
```

```python
rows=[]
_=db.apply(lambda row: [rows.append([row['types'], nn]) 
                         for nn in row.grapes], axis=1)
data_new=pd.DataFrame(rows)
data_new
```

```python
data_new[1]=[x.strip() for x in data_new[1]]
```

```python
data_new.rename(columns={0:"Type",1:"Grape"}, inplace=True)
```

```python
dt=data_new.groupby(by=["Type","Grape"]).size().copy()
```

```python
dt=pd.DataFrame(dt)
```

```python
dt.reset_index(inplace=True)
```

```python
dt.rename(columns={0:"Count"}, inplace=True)
```

```python
dt.to_csv("data/grapes_db.csv")
```
